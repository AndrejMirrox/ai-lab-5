Условие задачи: Определить скорость и расстояние, которое пройдет тело, падающее с высоты h = 100 метров с ускорением свободного падения g = 9.8 м/с^2.

Аналитическое решение поставленной задачи: Найдем время падения тела с высоты h. Для этого воспользуемся формулой свободного падения: h = 1/2 * g * t^2, где h - высота, с которой падает тело, g - ускорение свободного падения, t - время падения. Из этой формулы можно выразить время падения: t = sqrt(2 * h / g).

Найдем скорость падения тела. Для этого воспользуемся формулой скорости равноускоренного движения: v = g * t, где v - скорость тела, g - ускорение свободного падения, t - время падения.

Найдем расстояния, пройденные телом за каждую секунду падения. Для этого воспользуемся формулой расстояния равноускоренного движения: S = 1/2 * g * t^2, где S - расстояние, пройденное телом за время t, g - ускорение свободного падения.

 Программа для решения поставленной задачи:

```python
import matplotlib.pyplot as plt
import numpy as np

h = 100 # высота в метрах
g = 9.8 # ускорение свободного падения в м/с^2

# расчет времени падения
t = (2 * h / g) ** 0.5

# расчет скорости падения
v = g * t

# расчет расстояния, пройденного телом за каждую секунду падения
distances = [0.5 * g * i**2 for i in np.arange(int(t) + 1)]

# вывод результатов
print(f'Время падения: {t:.2f} секунд')
print(f'Скорость падения: {v:.2f} м/с')
print(f'Расстояния, пройденные телом за каждую секунду падения: {distances}')

# построение графика
plt.bar(np.arange(int(t) + 1), distances)
plt.title('Расстояния, пройденные телом за каждую секунду падения')
plt.xlabel('Время, с')
plt.ylabel('Расстояние, м')
plt.show()
```
