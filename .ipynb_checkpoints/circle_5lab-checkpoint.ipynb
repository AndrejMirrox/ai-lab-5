# Соотношение протеина и энергии в бургерах из МакДональдс (Взяты средние значения).

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import csv
import numpy as np
values=[]
with open('mcd.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    prot = []
    energ = []
    next(reader)
    for i in reader:
        prot.append(float(i[5]))
        energ.append(float(i[6]))
    prot_arr = np.array(prot)
    energ_arr = np.array(energ)
    values.append(np.mean(prot_arr))
    values.append(np.mean(energ_arr))
    plt.figure(figsize=(30, 30))
    explode = (0.1, 0.15)
    labels = ["Протеин","Энергия"]
    fig, ax = plt.subplots()
    ax.pie(values, labels=labels, shadow=True, explode=explode)
    plt.show()
```
