## Решение физической задачи с использованием линейного графика

Условие: В атмосфере находятся частицы пыли, имеющие массу m = 8 x 10^-22 кг и объем V = 5 x 10^-22 м3. Найти уменьшение их концентрации на высотах h_1 = 1.5 м и h_2 = 15 м. Считать, что воздух находится при нормальных условиях T_0 = 273 К, p_0 = 1.013 x 10^5 Па.

Аналитическое решение: Потенциальную энергию частицы с учетом архимедовой силы можно записать так: U(y)= (mg- pVg)y, где mg – сила тяжести;
pVg – выталкивающая сила Архимеда; p – плотность воздуха ( p = p0u / RT0 ), тогда расчетную формулу для отношения w = n / n0 концентраций запишем в виде (37) (рис. 3.6): w = exp(-e) , где
e = (m - pV)gy/kT .


```python
import matplotlib.pyplot as plt
import math as mt
import numpy as np
g=9.81;R=8.31;k=1.38e-23;mu0=0.029;T0=230;p0=1.0e5;
m=8.0e-22; V=5.0e-22; h1=1.5; h2=15;
ro=p0*mu0/(R*T0)

def dp(y):
    f=mt.exp(-(m-ro*V)*g*y/(k*T0));
    return f

print ("h1=",h1," w1=%7.3f"%dp(h1));
print ("h2=",h2," w2=%7.3f"%dp(h2));
ymin=0; ymax=25; N=100; dy=(ymax-ymin)/N;
z=[];DN=[]; z.append(ymin); DN.append(dp(ymin))

for i in np.arange(1,N,1):
    y1=ymin+i*dy; z.append(y1); DN.append(dp(y1))
 
plt.plot(z,DN,'k-')
plt.xlabel('$y$',fontsize=14)
plt.ylabel('$w$',fontsize=14)
plt.grid(True)
plt.show();
```
